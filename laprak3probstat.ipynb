```python
import pandas as pd

# Membaca data dari clipboard
df = pd.read_clipboard()

# Menampilkan jumlah total baris dalam dataframe
print(f"Jumlah total baris dalam dataframe: {len(df)}")

# Pastikan kolom "Waktu Perjalanan" ada di dataframe
if "Waktu Perjalanan" in df.columns:
    # Konversi ke format datetime atau timedelta (tergantung formatnya)
    try:
        df["Waktu Perjalanan"] = pd.to_timedelta(df["Waktu Perjalanan"])
    except:
        df["Waktu Perjalanan"] = pd.to_datetime(df["Waktu Perjalanan"], errors="coerce")

    # Sort dataframe berdasarkan kolom Waktu Perjalanan
    df_sorted = df.sort_values(by="Waktu Perjalanan", ascending=True)

    # Menyalin hasil ke clipboard
    df_sorted.to_clipboard(index=False)

    # Menampilkan hasil
    print(df_sorted)
else:
    print("⚠️ Kolom 'Waktu Perjalanan' tidak ditemukan dalam dataframe.")
```
